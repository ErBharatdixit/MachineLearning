```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python

corpus = []
for i in range(0,len(df)):
  review = re.sub('[^a-zA-Z]',' ',df['Message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True)
x = cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
y = pd.get_dummies(df['Category'])
y = y.iloc[:,1].values
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(x_train,y_train)
```

```python
# prediction
y_pred = spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
def avg_word2vec(doc):
  # Filter out words not in the model's vocabulary
  words_in_vocab = [word for word in doc if word in model.wv.index_to_key]
  if words_in_vocab:
    return np.mean([model.wv[word] for word in words_in_vocab], axis=0)
  else:
    # Return a zero vector of the model's vector size if no words are found in the vocabulary
    return np.zeros(model.vector_size)
```
